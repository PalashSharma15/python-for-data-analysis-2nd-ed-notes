##### Expressing Conditional Logic Array Operations

**np.where()** is numpy version of ternary expression **x if condition else y**

```
# Example: take a value from xarr, whenever the corresponding value of cond is True, else take value from yarr
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])

# list comprehension approach:
# not fast / not suitable for multi-dimensional arrays
result = [(x if c else y) for x, y, c in zip(xarr, yarr, cond)]
print(f'result from list comprehension: {result}')

# using np.where()
result = np.where(cond, xarr, yarr)
print(f'result from np.where(): {result}')
```

```
# Another example:
# to replace arguments can be scalar
# replace all positives with 2 and all negatives with -2
arr = np.random.randn(4, 4)
arr
```

#### 4.7 Example: Random Walks

Example Application for utilizing array operations. <br/>
Start: 0, Steps: 1 and -1 (equal probability)

```
# Pure python way to implement the same
import random
import matplotlib.pyplot as plt

position = 0
walk = [position]
steps = 1000
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)

plt.plot(walk[:100])
```

```
# Random walk with numpy
import numpy as np
import matplotlib.pyplot as plt

nsteps = 1000
draws = np.random.randint(0, 2, size = nsteps)

steps = np.where(draws > 0, 1, -1)
walk = steps.cumsum()
plt.plot(walk[:100])
```

```
walk.min()
```

```
walk.max()
```

```
# how long it took to get at least 10 steps away from the origin
(np.abs(walk) >= 10).argmax() # this will return the index
```

##### Simulating Many Random Walks at Once

```
import numpy as np
nwalks = 5000
nsteps = 1000
draws = np.random.randint(0, 2, size=(nwalks, nsteps)) # 0 or 1
steps = np.where(draws > 0, 1, -1)
walks = steps.cumsum(1) # since 0th axis is for 5000 individual walks
walks
```

```
walks.min()
```

```
walks.max()
```

```
# compute minimum crossing time to 30
hits30 = (np.abs(walks) >= 30).any(1)
hits30
```

```
# Number that hit 30 and -30
hits30.sum()
```

```
# Using boolean array to select out the rows of walks crossing absolute 30 level, and get crossing time
crossing_times = (np.abs(walks[hits30])).argmax(1) # extreacting indices
crossing_times
```

```
crossing_times.mean()
```
