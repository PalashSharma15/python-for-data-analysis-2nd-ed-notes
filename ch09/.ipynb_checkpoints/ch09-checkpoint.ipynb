###### Adjusting the spacing around the subplots

To change the default spacing, **subplots_adjust** method can be used on Figure objects.<br>
`subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)`
<br>**wspace** and **hspace** controls the percent of the figure width and height respectively.

```
# Example of no inter-subplot spacing
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True) # sharex and sharey when True, indicates subplotes have same x and y axes
for i in range(2):
  for j in range(2):
    axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
# matplotlib doesn't check whether axis labels will overlap or not (if it does, you can fix manually with tick locations and tick labels)
plt.subplots_adjust(wspace=0, hspace=0)
```

```
# Example of no inter-subplot spacing (with sharex and sharey to False, and with some wspace and hspace)
fig, axes = plt.subplots(2, 2, sharex=False, sharey=False)
for i in range(2):
  for j in range(2):
    axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=.5, hspace=.5)
```

###### Setting the title, axis labels, ticks, and ticklabels


```
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
ticks = ax.set_xticks([0, 250, 500, 750, 1000]) # setting x ticks
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small') # labels of xticks with 30 degree rotation and small font size
ax.set_title('My first matplotlib plot') # plot title
ax.set_xlabel('Stages') # x axis label
```

```
# a more easy way is using set method
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
ticks = ax.set_xticks([0, 250, 500, 750, 1000]) # setting x ticks
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small') # labels of xticks with 30 degree rotation and small font size
props = {
    'title': 'My first matplotlib plot',
    'xlabel': 'Stages',
    'ylabel': 'Y'
}
ax.set(**props)
```

##### Annotations and Drawing a Subplot


```
from datetime import datetime
import pandas as pd
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv('https://raw.githubusercontent.com/wesm/pydata-book/2nd-edition/examples/spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']

spx.plot(ax = ax, style='k-')

crisis_data = [
               (datetime(2007, 10, 11), 'Peak of bull market'),
               (datetime(2008, 3, 12), 'Bear Stearns Fails'),
               (datetime(2008, 9, 15), 'Lehman Bankrupty')
]
for date, label in crisis_data:
  ax.annotate(label, xy=(date, spx.asof(date) + 75), # xy is the point to annotate
              xytext=(date, spx.asof(date) + 225), # location of label text
              arrowprops = dict(facecolor='black', headwidth=4,
                                width=2, headlength=4),
              horizontalalignment = 'left', verticalalignment='top') # to keep text within the plot

# Zoom on 2007-2010 (setting boundaries)
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])
ax.set_title('Important dates in the 2008-2009 financial crisis')
```

```
ax.annotate?
```

Drawing shapes with matplotlib<br>
1. matplotlib has objects that represents many common shapes, referred to as **patches**<br>
2. Some common shapes like Rectangle and Circle are found in **matplotlib.pyplot**<br>
3. full set is located in **matplotlib.patches**

```
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# 1. add a shape to a plot (creating patch)
rect = plt.Rectangle(xy=(0.2, 0.75), width=0.4, height=0.15, color='k', alpha=0.3)
circ = plt.Circle(xy=(0.7, 0.2), radius=0.15, color='b', alpha=0.3)
pgon = plt.Polygon(xy=[[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color='g', alpha=0.5) # xy is a numpy array of shape N x 2, where N is the number of points
# 2. add it to a subplot by calling add_patch() method
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

DataFrame's plot method plots each of its columns as a different line on the same subplot

```
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
                  columns=['A', 'B', 'C', 'D'],
                  index = np.arange(0, 100, 10))
df.plot() # creates legend automatically (df.plot() is equivalent to df.plot.line())
```

##### Bar Plots

1. plot.bar() -> vertical bars
2. plot.barh() -> horizontal bars

```
fig, axes = plt.subplots(2, 1) # defining 2 subplots
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop')) # data frame
data.plot.bar(ax = axes[0], color='k', alpha=0.7) # plotting vertical bars on ax[0]
data.plot.barh(ax = axes[1], color='k', alpha=0.7, figsize=(8, 8)) # plotting horizontal bars on ax[1]
```

More about Stacked bar plot:

```
tips = pd.read_csv("https://raw.githubusercontent.com/wesm/pydata-book/2nd-edition/examples/tips.csv")
tips.head()
```

```
party_counts = pd.crosstab(tips['day'], tips['size']) # calculating frequency on the basis of day and size column
party_counts
```

```
party_counts = party_counts.loc[:, 2:5] # not many 1 and 6 person parties
# normalize so that each row sums to 1
party_pcts = party_counts.div(party_counts.sum(1), axis=0)
party_pcts.plot.bar() # Fraction of parties by size on each day
```

Using seaborn:

```
import seaborn as sns
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips.head()
```

```
print('black lines represents 95% confidence interval')
# input can be a dataframe with specified column names
sns.barplot(x='tip_pct', y='day', data=tips, orient='h').set_title('tipping percentage by day with error bars')
```

**hue** option in sns.barplot()

```
sns.barplot(x='tip_pct', y='day', hue='time', data=tips, orient='h').set_title('Tipping percentage by day and time')
```

Switch between different plot appearances using **sns.set()**:

```
sns.set(style='whitegrid')
sns.barplot(x='tip_pct', y='day', hue='time', data=tips, orient='h').set_title('Tipping percentage by day and time')
```

Plotting both histogram and density plot simultaneously with **sns.distplot()** method:

```
# Consider a bimodal distribution consisting of draws from two different standard normal distribution
comp1 = np.random.normal(0, 1, size=200)
comp2 = np.random.normal(10, 2, size=200)
values = pd.Series(np.concatenate([comp1, comp2]))
sns.distplot(values, bins=100, color='k').set_title('Normalized histogram of normal mixture with density estimate')
```

##### Scatter or Point Plots

Used for examining 2 One-Dimensional data series

```
# select a few variables from macro dataset and then, compute log difference
macro = pd.read_csv('https://raw.githubusercontent.com/wesm/pydata-book/2nd-edition/examples/macrodata.csv')
```

```
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data = np.log(data).diff().dropna()
trans_data[-5:]
```

seaborn's **regplot** method makes a scatter plot and fits a linear regression line

```
sns.regplot(x='m1', y='unemp', data=trans_data)
plt.title('Changes in log %s versus log %s' % ('m1', 'unemp'))
```

**Scatter Plot Matrix** or **Pairs Plot**: scatter plots for a group of varaibles <br>
**pariplot** function: supports placing histogram or density estimates of each variable along the diagonal

```
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha': 0.2})
plt.suptitle('Pair plot matrix of statesmodel macro data') # Why plt.suptitle()? -> suptitle stands for supertitle (for overall figure)
```

##### Facet Grids and Categorical Data

Tipping percentage by day/time/smoker:

```
# hue by 'time', and differentiate both on the basis of 'smoker'
sns.factorplot(x='day', y='tip_pct', hue='time', col='smoker', kind='bar', data=tips[tips.tip_pct < 1])
```

We can also expand the facet grid

```
# tip_pct by day; facet by time / smoker
sns.factorplot(x='day', y='tip_pct', row='time', col='smoker', kind='bar',
               data=tips[tips.tip_pct < 1])
```

**factorplot** supports other plot types also.<br>
Example of box plot using factorplot:

```
sns.factorplot(x='day', y='tip_pct', kind='box', data=tips[tips.tip_pct < 0.5]) # see seaborn.FacetGrid class for more facet grid plots
```
