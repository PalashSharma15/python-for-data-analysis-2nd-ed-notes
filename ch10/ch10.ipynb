# 10.1 GroupBy Machenics

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],
                   'key2' : ['one', 'two', 'one', 'two', 'one'],
                   'data1' : np.random.randn(5),
                   'data2' : np.random.randn(5)})
df
```

Compute the mean of the data1 column using key1 labels

```python
grouped = df['data1'].groupby(df['key1'])
grouped # till now, only computed intermediate data about the key1
```

```python
# computing mean
grouped.mean()
```

```python
# if we pass multiple series as a list in the key option
means = df['data1'].groupby([df['key1'], df['key2']]).mean()
means
```

```python
# if we unstack
means.unstack()
```

```python
# passing arrays instead of series in the group by key
states = np.array(['Ohio', 'California', 'California', 'Ohio', 'Ohio'])
years = np.array([2005, 2005, 2006, 2005, 2006])
df['data1'].groupby([states, years]).mean()
```

```python
# if we need to groupby based on the columns of dataframe itslef
df.groupby('key1').mean()
```

```python
# with multiple keys
df.groupby(['key1', 'key2']).mean()
```

```python
# size method returns Series containing group sizes regardless of the datatype
df.groupby(['key1', 'key2']).size()
```

## Iterating Over Groups

```python
for name, group in df.groupby('key1'):
    print(name) # group name
    print(group) # data
```

```python
# iterating with multiple keys
for (k1, k2), group in df.groupby(['key1', 'key2']):
    print(k1, k2)
    print(group)
```

```python
# creating a data dictionary using groupby
pieces=dict(list(df.groupby('key1')))
pieces['b']
```

```python
# default groupby axis is 0
# we can modify that
df.dtypes
```

```python
grouped = df.groupby(df.dtypes, axis=1)
```

```python
for dtype, group in grouped:
    print(dtype)
    print(group)
```

## Grouping with Dicts and Series

```python
people = pd.DataFrame(np.random.randn(5, 5), columns=list('abcde'),
                      index=['Joe', 'Steve', 'Wes', 'Jim', 'Travis'])
people.iloc[2:3, [1, 2]] = np.nan # NA values
people
```

```python
# sum together columns by group
mapping = {'a': 'red', 
          'b': 'red',
          'c': 'blue',
          'd': 'blue',
          'e': 'red',
          'f': 'orange'} # f is not present in the columns and hence it will be ignored by groupby
```

```python
by_column = people.groupby(mapping, axis=1)
by_column.sum()
```

```python
# instead of dictionary we can also pass a series of mapped column names
map_series = pd.Series(mapping)
people.groupby(map_series, axis=1).count()
```

## Grouping with functions

```python
# group by indices length
people.groupby(len).sum()
```

```python
# we can mix functions, arrays, dicts together
key_list = ['one', 'one', 'one', 'two', 'two']
people.groupby([len, key_list]).min()
```

# 10.3 Apply: General split-apply-combine 

```python
# selecting top n tip_pct values by group
def top(df, n=5, column='tips_pct'):
    return df.sort_values(by=column)[-n:]

top(tips, n=6)
```

```python
# using apply method
tips.groupby('smoker').apply(top)
```

```python
# passing arguments for the function within apply method
tips.groupby(['smoker', 'day']).apply(top, n=1, column='total_bill')
```

```python
# one way to use describe with groupby
result = tips.groupby('smoker')['tips_pct'].describe()
result
```

```python
result.unstack()
```

```python
# which is similar to this approach
f = lambda x: x.describe()
tips.groupby('smoker')['tips_pct'].apply(f)
```

## Supressing the Group Keys 

```python
tips.groupby('smoker').apply(top) # with group keys
```

```python
# without group keys
tips.groupby('smoker', group_keys=False).apply(top)
```

## Quantile and Bucket Analysis

```python
frame = pd.DataFrame({'data1': np.random.randn(1000), 'data2': np.random.randn(1000)})
quartiles = pd.cut(frame.data1, 4)
quartiles[:10]
```

```python
# these categorical object can be passed to groupby
def get_stats(group):
    return {'min': group.min(), 'max': group.max(),
           'count': group.count(), 'mean': group.mean()}
grouped = frame.data2.groupby(quartiles)
grouped.apply(get_stats).unstack()
```

```python
# pass label = False for just numbers
grouping = pd.qcut(frame.data1, 10, labels=False)
grouped = frame.data2.groupby(grouping)
grouped.apply(get_stats).unstack()
```

## Example: Filling Missing  Values with Group-Specific Values

```python
s = pd.Series(np.random.randn(6))
s[::2] = np.nan
s
```

```python
# filling NA with mean
s.fillna(s.mean())
```

```python
# fill value vary by group
states = ['Ohio', 'New York', 'Vermont', 'Florida','Oregon', 'Nevada', 'California', 'Idaho']
group_key = ['East'] * 4 + ['West'] * 4
```

```python
data = pd.Series(np.random.randn(8), index = states)
data
```

```python
# assign null values
data[['Vermont', 'Nevada', 'Idaho']] = np.nan
data
```

```python
data.groupby(group_key).mean()
```

```python
# Filling missing values using group means
fill_mean = lambda g: g.fillna(g.mean())
data.groupby(group_key).apply(fill_mean)
```

```python
# using predefined values
fill_values = {'East': 0.5,
              'West': -1}
fill_func = lambda g: g.fillna(fill_values[g.name])
data.groupby(group_key).apply(fill_func)
```

## Example: Random Sampling and Permutation

```python
# Using playing cards deck example
# Hearts,  Spades, Clubs, Diamond

suits = list('HSCD')
card_val = (list(range(1, 11)) + [10] * 3) * 4
base_names = ['A'] + list(range(2, 11)) + list('JKQ')
cards = []
for suit in suits:
    cards.extend(str(num) + suit for num in base_names)

deck = pd.Series(card_val, index=cards)
deck[:13]
```

```python
# function to draw a hand of five cards from the deck
def draw(deck, n=5):
    return deck.sample(n)

draw(deck)
```

```python
# drawing two random cards from each suit
get_suit = lambda card: card[-1] # last letter = suit
deck.groupby(get_suit).apply(draw, n=2)
```

```python
# Alternative approach
deck.groupby(get_suit, group_keys=False).apply(draw, n=2)
```

## Example: Group Weighted Average Correlation

```python
df = pd.DataFrame({'category': list('aaaabbbb'),
                   'data': np.random.randn(8), 'weights': np.random.rand(8)})
df
```

```python
grouped = df.groupby('category')
get_wavg = lambda g: np.average(g['data'], weights=g['weights'])
grouped.apply(get_wavg)
```

Compute a dataframe consisting of the yearly correlations of daily returns computed from prcnt changes with SPX from Yahoo Finance dataset

```python
close_px = pd.read_csv('../examples/stock_px_2.csv',
                      parse_dates=True, index_col =  0)
close_px.info()
```

```python
close_px[-5:]
```

```python
# function for pairwise correlation calculation
spx_corr = lambda x: x.corrwith(x['SPX'])
# calculate pct change
rets = close_px.pct_change().dropna()
# grouping pct change by years
get_year = lambda x: x.year
by_year = rets.groupby(get_year)
by_year.apply(spx_corr)
```

```python
# inter column correlations
by_year.apply(lambda g: g['AAPL'].corr(g['MSFT']))
```

## Example: Group-Wise Linear Regression

```python
# defining an OLS regression on each chunk of data
import statsmodels.api as sm
def regress(data, yvar, xvars):
    Y = data[yvar]
    X = data[xvars]
    X['intercept'] = 1
    result = sm.OLS(Y, X).fit()
    return result.params
```

```python
# applying linear regression on yearly data
by_year.apply(regress, 'AAPL', ['SPX'])
```

## Cross-Tabulations: Crosstab

    Special case of pivot-table that computes frequencies

```python
data = pd.DataFrame({'Sample':list(range(1, 11)),
                    'Nationality':['USA', 'Japan', 'USA', 'Japan', 'Japan', 'Japan',
                                  'USA', 'USA', 'Japan', 'USA'],
                    'Handedness': [hnd+'-handed' for hnd in ['Right', 'Left', 'Right', 'Right', 'Left', 'Right', 'Right',
                                   'Left', 'Right', 'Right']]})
data
```

```python
# summarise data by nationality and handednes
pd.crosstab(data.Nationality, data.Handedness, margins=True)
```

```python
# first two arguments can either be series or arrays or list
pd.crosstab([tips.time, tips.day], tips.smoker, margins=True)
```
